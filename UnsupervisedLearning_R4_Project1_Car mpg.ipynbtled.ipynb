```python
numeric_cols = cars_df.drop('car name', axis=1)

# Copy the 'mpg' column alone into the y dataframe. This is the dependent variable
car_names = pd.DataFrame(cars_df[['car name']])


numeric_cols = numeric_cols.apply(lambda x: x.fillna(x.median()),axis=0)
cars_df = numeric_cols.join(car_names)   # Recreating mpg_df by combining numerical columns with car names

cars_df.info()
```

```python
#Let us do a pair plot analysis to visually check number of likely clusters
```

```python
# This is done using scatter matrix function which creates a dashboard reflecting useful information about the dimensions
# The result can be stored as a .png file and opened in say, paint to get a larger view 
import seaborn as sns
cars_df_attr = cars_df.iloc[:, 0:9]
cars_df_attr['dispercyl'] = cars_df_attr['disp'] / cars_df_attr['cyl']
sns.pairplot(cars_df_attr, diag_kind='kde')   # to plot density curve instead of histogram

# hue = where we n eed to colour byh this 
#sns.pairplot(mpg_df_attr)  # to plot histogram, the default

# multiple peaks are not seen for the accelartion.
```

```python
from scipy.stats import zscore
# cars_df_attr = cars_df_attr.drop('car name', axis=1)
# cars_df_attr = cars_df.loc[:, 'mpg':'origin']
cars_df_attr_z = cars_df_attr.apply(zscore)

     # Remove "origin" and "yr" columns
cars_df_attr_z.pop('yr')

array = cars_df_attr_z.values
# cars_df_attr = cars_df_attr.drop('car name', axis=1)
```

```python
from sklearn.cluster import KMeans
cluster_range = range( 2, 6)   # expect 3 to four clusters from the pair panel visual inspection hence restricting from 2 to 6
cluster_errors = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 5)
  clusters.fit(cars_df_attr)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
clusters_df[0:15]
```

```python
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```
