# Q5 Remove any unwanted columns or outliers, standardize variables in pre-processing step

```python
# We could see most of the outliers are now removed.


# here remoe=ved ID column as it does not have any effect on the result
from sklearn.cluster import KMeans
cluster_range = range( 2, 10)   # expect 3 to four clusters from the pair panel visual inspection hence restricting from 2 to 6
cluster_errors = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 5)
  clusters.fit(bc_df)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
print(clusters_df[0:15])

# Elbow plot

plt.figure(figsize=(12,10))
plt.plot(clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )


# here number of clusters comes out to be 5 with only scaling we can see .

```

# Q6 Create a covariance matrix for identifying Principal components

```python
# PCA
# Step 1 - Create covariance matrix


from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(bc_df)


cov_matrix = np.cov(X_std.T)
print('Covariance Matrix \n%s', cov_matrix)
```

# Q7 Identify eigen values and eigen vector

```python
# Step 2- Get eigen values and eigen vector
```

```python
# USing PCA  to normalize the data

#variance-covariance structure of a set of variables through linear combinations. It is often used as a dimensionality-reduction technique.




#Centering the Data

from sklearn.decomposition import PCA
X_centered = bc_df - bc_df.mean(axis=0)

X_centered

pca = PCA(n_components=10)
pca.fit(X_centered)


X_pca = pca.transform(X_centered)
print(X_pca.shape)
print(X_pca)
print(' pca components eigen vectors',pca.components_)
print( 'explained variance', pca.explained_variance_ratio_)
```
